```python
#import the necessary libraries
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#read the image
img = cv2.imread('Lenna.png',0)
```

```python
#display image
plt.imshow(img, 'gray')
plt.title('my picture')
plt.show()
```

```python
#define filters
sobelx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype = np.float)
sobely = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype = np.float)
```

```python
#get vertical edges
gx = cv2.filter2D(img, -1, sobelx)
```

```python
#get horizontal edges
gy = cv2.filter2D(img, -1, sobely)
```

```python
#cast between 0 and 255
g_x = gx* 255.0 / np.max(gx)
g_x = g_x.astype(np.uint8)
plt.figure('2D Convolution Gradient')
plt.imshow(g_x, cmap=plt.cm.gray)   
```

```python
#cast between 0 and 255
g_y = gy * 255.0 / np.max(gy)
g_y = g_y.astype(np.uint8)
plt.figure('2D Convolution Gradient vertical')
plt.imshow(g_y, cmap=plt.cm.gray)   
```

```python
#get the magnitude of the edges
g = np.hypot(gx,gy)
g *= 255.0 / np.max(g)
g = g.astype(np.uint8)
plt.figure('2D Convolution Gradient horizontal')
plt.imshow(g, cmap=plt.cm.gray)   
```

```python
theta = np.arctan2(gx, gy)
```

```python
def non_max_suppression(img, D):
    M, N = img.shape
    Z = np.zeros((M,N), dtype=np.int32)
    angle = D * 180. / np.pi
    angle[angle < 0] += 180

    #iterate through each pixel
    for i in range(1,M-1):
        for j in range(1,N-1):
            try:
                q = 255
                r = 255
                
               #angle 0
                if (0 <= angle[i,j] < 22.5) or (157.5 <= angle[i,j] <= 180):
                    q = img[i, j+1]
                    r = img[i, j-1]
                #angle 45
                elif (22.5 <= angle[i,j] < 67.5):
                    q = img[i+1, j-1]
                    r = img[i-1, j+1]
                #angle 90
                elif (67.5 <= angle[i,j] < 112.5):
                    q = img[i+1, j]
                    r = img[i-1, j]
                #angle 135
                elif (112.5 <= angle[i,j] < 157.5):
                    q = img[i-1, j-1]
                    r = img[i+1, j+1]

                if (img[i,j] >= q) and (img[i,j] >= r):
                    Z[i,j] = img[i,j]
                else:
                    Z[i,j] = 0

            except IndexError as e:
                pass
    
    return Z
```

```python
final_img = non_max_suppression(g, theta)
```

```python
plt.figure('2D Convolution Gradient horizontal')
plt.imshow(final_img, cmap=plt.cm.gray)   
```
