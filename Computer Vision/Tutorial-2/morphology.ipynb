```python
#import libarites
import cv2
import numpy as np
from matplotlib import pyplot as plt

#read and show image
img = cv2.imread('circuit.jpg',0)
plt.imshow(img,'gray')
plt.title('my pict')
plt.show()
```

```python
#Threshold the image
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
#display image
plt.imshow(thresh1,'gray')
plt.title('my pict')
plt.show()
```

```python
#dilation
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(thresh1,kernel,iterations = 1)
```

```python
plt.imshow(dilation,'gray')
plt.title('my pict')
plt.show()
```

```python
#erosion
kernel = np.ones((8,8),np.uint8)
erotion = cv2.erode(thresh1,kernel,iterations = 1)
```

```python
plt.imshow(erotion,'gray')
plt.title('my pict')
plt.show()
```
